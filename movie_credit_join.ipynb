```python
from pyspark.sql import SparkSession

# Initialize a Spark session
spark = SparkSession.builder.appName("MovieDataJoin").getOrCreate()
```

```python
# Load the CSV files into Spark DataFrames
director_dummies_df = spark.read.csv("Resources/director_dummies.csv", header=True, inferSchema=True)
dummy_moviedata_df = spark.read.csv("Resources/dummy_moviedata_df.csv", header=True, inferSchema=True)
```

```python
dummy_moviedata_df.show(5)
```

```python
director_dummies_df.show(5)
```

```python
# Join the DataFrames using the 'movie_id' column
joined_df = dummy_moviedata_df.join(director_dummies_df, on="movie_id")
joined_df = joined_df.drop('_c0')
joined_df = joined_df.drop('Movie2')
joined_df.show(5)
```

```python
# Write the joined DataFrame to a CSV file
joined_df.coalesce(1).write.csv("joined_movie_data", header=True)
```

```python
# Stop the Spark session
spark.stop()
```
