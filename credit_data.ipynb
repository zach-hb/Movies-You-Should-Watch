```python
import pandas as pd
import ast
```

```python
movies_df = pd.read_csv('Resources/tmdb_5000_movies.csv')
movies_df.head(1)
```

```python
credits_df = pd.read_csv('Resources/tmdb_5000_credits.csv')
credits_df.head(1)
```

Extracting each director

```python
credits_df['crew'][1]
```

```python
credits_df.info()
```

```python
# checks if each row is a string, then changes it into an object 
credits_df['crew'] = credits_df['crew'].apply(lambda x: ast.literal_eval(x) if isinstance(x, str) else x)
```

```python
directors_list = []

for row in credits_df['crew']:
    for i in row:
        if i.get('job') == 'Director':
            directors_list.append(i.get('name'))
            break
    else:
        directors_list.append(None)    
directors_list[:5]
```

```python
directors_df = pd.DataFrame(directors_list, columns=['Director'])
directors_df.tail()
```

Converting each director to its own column using pd.get_dummies()

```python
dummy_director_df = pd.get_dummies(directors_df[['Director']])
dummy_director_df = dummy_director_df.astype(int)
dummy_director_df.columns = dummy_director_df.columns.str.replace('Director_', '')
dummy_director_df.head()
```

```python
dummy_director_df[['movie_id', 'title']] = credits_df[['movie_id','title']]
dummy_director_df.head()
```

Exporting to .csv!

```python
dummy_director_df.to_csv('Resources/director_dummies.csv')
```
