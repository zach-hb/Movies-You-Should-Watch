```python
# Import Dependencies
import requests
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

```

```python
# Read the csv files
movie_data = pd.read_csv('../Movies-You-Should-Watch/Resources/tmdb_5000_movies.csv')
movie_credits = pd.read_csv('../Movies-You-Should-Watch/Resources/tmdb_5000_credits.csv')
movie_credits.head()
movie_data.head()
```

```python
# Merge the datasets
movie_credits.columns = ['id', 'title', 'cast', 'crew']
movie_data = movie_data.merge(movie_credits, on="id")
movie_data.head()
```

```python
# convert string representations of lists in actual lists
from ast import literal_eval

features = ["cast", "crew", "keywords", "genres"]

for feature in features:
    movie_data[feature] = movie_data[feature].apply(literal_eval)

# Display the first few rows of each modified feature column
for feature in features:
    print(f"{feature}:")
    print(movie_data[feature].head(10))
    print("\n")
```

```python
print(movie_data.columns)
```

```python
# Extract the director of the movie
def get_director(x): 
    for i in x: 
        if i["job"] =="Director": 
            return i["name"]
    return np.nan

# extract list of names for each feature
def get_list(x): 
    try: 
        if isinstance(x, list): 
            names = [i["name"] for i in x]

            if len(names) > 3: 
                names = names[:3]
            return names
        return[]
    except (KeyError, TypeError): 
        return[]

# apply both functions get_director() and get_list()
movie_data["director"] = movie_data["crew"].apply(get_director)

features = ["cast", "keywords", "genres"]
for feature in features: 
    movie_data[feature] = movie_data[feature].apply(get_list)

movie_data[['original_title', 'cast', 'director', 'keywords', 'genres']].head()
```

```python
# Create "soup" column by combining info from different columns into a single string
def create_soup(features):
    keywords = ' '.join(features['keywords']) if isinstance(features['keywords'], list) else ''
    cast = ' '.join(features['cast']) if isinstance(features['cast'], list) else ''
    director = str(features['director']) if pd.notna(features['director']) else ''
    genres = ' '.join(features['genres']) if isinstance(features['genres'], list) else ''
    
    return keywords + ' ' + cast + ' ' + director + ' ' + genres

movie_data["soup"] = movie_data.apply(create_soup, axis=1)
print(movie_data["soup"].head())
```

```python

# Prepare data and calculate cosine similarity scores between movies
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity

count_vectorizer = CountVectorizer(stop_words="english")
count_matrix = count_vectorizer.fit_transform(movie_data["soup"])

print(count_matrix.shape)

cosine_sim2 = cosine_similarity(count_matrix, count_matrix) 
print(cosine_sim2.shape)

movie_data = movie_data.reset_index()
indices = pd.Series(movie_data.index, index=movie_data['original_title'])
```

```python
# Reverse mapping from movie titles to corresponding indices in movie_data 
indices = pd.Series(movie_data.index,
index=movie_data["original_title"]).drop_duplicates()

print(indices.head())
```

```python
# Implement the content-based recommendation system using cosine similarity
def get_recommendations(title, cosine_sim=cosine_sim2):
    idx = indices[title]
    similarity_scores = list(enumerate(cosine_sim[idx]))
    similarity_scores = sorted(similarity_scores, key=lambda x: x[1], reverse=True)
    similarity_scores = similarity_scores[1:11]  # Corrected variable name
    # (a, b) where a is id of movie, b is similarity score

    movie_indices = [ind[0] for ind in similarity_scores]
    recommended_movies = movie_data["original_title"].iloc[movie_indices]
    return recommended_movies

print("################ Content Based System #############")
print("Recommendations for The Dark Knight Rises:")
print(get_recommendations("The Dark Knight Rises"))
print()
print("Recommendations for Avengers:")
print(get_recommendations("The Avengers"))
```
